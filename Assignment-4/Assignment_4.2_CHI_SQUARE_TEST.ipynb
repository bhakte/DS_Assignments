Association between Device Type and Customer Satisfaction

**Background:->**Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices: Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the type of device purchased and the customer's satisfaction level.

Data Provided:
The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices:

image.png

**Objective:->**To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.

**Assignment Tasks:**
1. State the Hypotheses:
2. Compute the Chi-Square Statistic:
3. Determine the Critical Value:
Using the significance level (alpha) of 0.05 and the degrees of freedom (which is the number of categories minus 1)
4. Make a Decision:
Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.




```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency, chi2
```

1.   **State the Hypotheses**

**Null Hypothesis (H₀):->**
There is no association between device type and customer satisfaction level.
(i.e., they are independent)

**Alternative Hypothesis (H₁):->**
There is an association between device type and customer satisfaction level.
(i.e., they are not independent)

```python
# Step 1: Create the observed frequency table (contingency table)
# Each row represents a satisfaction level, and each column represents a device type
data = np.array([
    [50, 70],  # Very Satisfied
    [80, 100], # Satisfied
    [60, 90],  # Neutral
    [30, 50],  # Unsatisfied
    [20, 50]   # Very Unsatisfied
])
```

```python
# Creating labels for better readability
index_labels = ['Very Satisfied', 'Satisfied', 'Neutral', 'Unsatisfied', 'Very Unsatisfied']
column_labels = ['Smart Thermostat', 'Smart Light']
```

```python
# Create a pandas DataFrame for display and calculations
observed_df = pd.DataFrame(data, index=index_labels, columns=column_labels)
print("Observed Frequency Table:\n")
print(observed_df)
```

***Note : to Compute Expected Frequencies manually***

image.png


```python
# Step 2: Perform the Chi-Square Test using scipy
chi2_stat, p_val, dof, expected = chi2_contingency(data)
chi2_stat, p_val, dof, expected
```

```python
print("\nExpected Frequency Table:\n")
expected_df = pd.DataFrame(expected, index=index_labels, columns=column_labels)
print(expected_df)
```

```python
# Step 3: Show the chi-square statistic and degrees of freedom
print(f"\nChi-Square Statistic = {chi2_stat:.3f}")
print(f"Degrees of Freedom = {dof}") #formula for Degrees of Freedom -> df=(r−1)⋅(c−1)=(5−1)(2−1)=4. where r-row count, c-column count
```

```python
# Step 4: Find the critical value from chi-square distribution table
alpha = 0.05  # Significance level
critical_value = chi2.ppf(q=1 - alpha, df=dof)
print(f"Critical Value at α = {alpha} and df = {dof} is {critical_value:.3f}")
```

```python
# Step 5: Make a decision
if chi2_stat > critical_value:
    print("\nResult: Reject the null hypothesis.")
    print("Conclusion: There is a significant association between device type and customer satisfaction.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("Conclusion: There is no significant association between device type and customer satisfaction.")
```
