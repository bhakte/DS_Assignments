**Note** : All columns are numeric (float64)  no need for type conversion.

```python
# Drop all rows that contain any missing values
df_clean = data_orig.dropna() #Just ~1% of the data is missing  very low impact if dropped. i.e only 21 rows
```

```python
# Check how many rows remain
print(f"Rows before deletion: {data_orig.shape[0]}")
print(f"Rows after deletion: {df_clean.shape[0]}") #After droping all rows that contain any missing values. 2105 rows remain
```

```python
# Use IQR method to remove outliers
Q1 = df_clean.quantile(0.25)
Q3 = df_clean.quantile(0.75)
IQR = Q3 - Q1
#Q1, Q3, IQR
```

```python
# Filter for rows within 1.5 * IQR
df_no_outliers = df_clean[
    ~((df_clean < (Q1 - 1.5 * IQR)) | (df_clean > (Q3 + 1.5 * IQR))).any(axis=1)
]
```

```python
print(f"\nOriginal Rows: {data_orig.shape[0]}")
print(f"After Cleaning & Outlier Removal: {df_no_outliers.shape[0]}")
```

```python
# Generate a statistical summary (count, mean, std, min, 25%, 50%, 75%, max) of the dataframe without outliers
summary_stats = df_no_outliers.describe().T
summary_stats["median"] = df_no_outliers.median() # Add a new column "median" to the summary by computing the median for each column in the dataframe
summary_stats["IQR"] = summary_stats["75%"] - summary_stats["25%"] # Add a new column "IQR" (Interquartile Range) calculated as the difference between 75th percentile and 25th percentile
print("\nStatistical Summary:") # Print the final summary statistics table including mean, std, median, IQR, etc.
print(summary_stats)
```

**Highlight any interesting findings from this summary.**
1. LB (Baseline Fetal Heart Rate):
*   Mean: ~131.85 bpm
*   Range: 106 to 160 bpm
*  IQR (Interquartile Range): 13
*   Most values are tightly clustered around the median of 133, with low standard deviation (9.45)  indicating consistency in fetal heart rate across observations.
2. AC, FM, UC, DL (Short-term Fetal Movements and Signal Features):


*   These features (especially AC, FM, DL) have very small means and standard deviations and many zero values, suggesting:

    -> Sparsity: Most values are near 0 (e.g., FM, DL have a median of 0 and very low IQR).

    -> Potential outliers (small negative minimums like FM = -0.0024, ALTV = -1.04) which might be artifacts or noise.

3. DS and DP (Possibly Discrete/Boolean features):


*   All values are exactly zero.
    - > These may be redundant or unused variables, and candidates for removal in modeling unless domain knowledge suggests otherwise.

4. ASTV & MSTV (Signal Variability Measures):


*   ASTV (Accelerations per second):
    - > Wide variability: mean = 42.39, std = 15.08
    - >Range: 15 to 79
    ->Indicates variation in fetal heart accelerations, which may be meaningful for health classification.
*   MSTV (Short term variability):
    - >Mean = 1.27, low std = 0.53
    ->Tight distribution, indicating consistent variability across subjects.

5. ALTV & MLTV (Long-term variability metrics):
   - > ALTV has high skewness (mean = 3.16, median = 0, max = 27)  most values are 0, but some extreme long-term changes
   - > MLTV shows a more normal distribution with moderate spread (std = 4.49).

6. Width:
   - > High spread: std = 32.58, max = 176.3
   - > This could be a significant feature  possibly representing the width of signal windows  but the large variance suggests it might need normalization.
7. Tendency:
    - > Range from -1.00 to 1.02 with mean close to 0
    - > Looks like a trend or directional feature (e.g., increasing/decreasing signal).
    - > Median = 0, IQR = 1: shows balanced tendency overall.

8. NSP (Target Variable):
    - > All values are 1, with zero standard deviation
        - > Important: This suggests there's no class variation  making it unsuitable for classification unless the full dataset has more class diversity. We may be looking at a filtered or subset version.

```python
#bar charts or pie charts to display the frequency of categories for categorical variables.
#  Step 1: Value counts of categorical column (NSP)
category_counts = df_clean['NSP'].value_counts().sort_index()

#  Step 2: Bar Chart
plt.bar(category_counts.index, category_counts.values, color='skyblue')
plt.xlabel("NSP Class")
plt.ylabel("Frequency")
plt.title("Frequency of NSP (Fetal States)")
plt.xticks([1, 2, 3], ['Normal (1)', 'Suspect (2)', 'Pathological (3)'],rotation=15 )
plt.tight_layout()
plt.show()
```

```python
#  Violin Plot (Density & Distribution)
# -----------------------------------------------
plt.figure(figsize=(14, 6))
sns.violinplot(data=df_no_outliers[selected_features], inner='box', palette="Set2")
plt.title("Violin Plot of Selected Features")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
