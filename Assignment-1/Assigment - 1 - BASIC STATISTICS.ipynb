**Standardization of Numerical Variables**

To scale numerical variables for uniformity, improving the datasetâ€™s suitability for analytical models.

	Explain the concept of standardization (z-score normalization).
Ans-> Standardization, also known as Z-score normalization, is a technique used to rescale data so that it has a mean of 0 and a standard deviation of 1. This transformation makes different features comparable and is particularly useful for machine learning algorithms that rely on distance measurements, such as k-NN, SVM, and logistic regression.

Why Use Standardization?
->


1.   Handles Different Scales â†’ Some features might have large values (e.g., sales in thousands), while others are small (e.g., discount rates as fractions). Standardization makes them comparable.
2.   Speeds Up Training â†’ Many ML algorithms converge faster when features have a similar scale.
3.   Improves Accuracy â†’ Algorithms that use distance-based metrics (e.g., k-NN, K-Means) perform better with standardized data.

```python
# Identify numerical columns
numerical_columns7 = df.select_dtypes(include=['int64', 'float64']).columns
numerical_columns7
```

```python
# Compute mean and standard deviation manually
mean_values = df[numerical_columns7].mean()
mean_values
```

```python
std_values = df[numerical_columns7].std()
std_values
```

```python
# Apply Z-score standardization formula: Z = (X - Î¼) / Ïƒ
df_standardized = df.copy()
df_standardized[numerical_columns7] = (df[numerical_columns7] - mean_values) / std_values
```

```python
# Plot before and after distributions for comparison
for col in numerical_columns7:
    plt.figure(figsize=(10, 4))

    # Original distribution
    plt.subplot(1, 2, 1)
    plt.hist(df[col], bins=20, color='blue', alpha=0.6, edgecolor='black')
    plt.title(f'Original Distribution: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Standardized distribution
    plt.subplot(1, 2, 2)
    plt.hist(df_standardized[col], bins=20, color='green', alpha=0.6, edgecolor='black')
    plt.title(f'Standardized Distribution: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```

**Conversion of Categorical Data into Dummy Variables**

->To transform categorical variables into a format that can be provided to ML algorithms.

**Why Convert Categorical Data into Dummy Variables (One-Hot Encoding)?**
->Many machine learning models and statistical algorithms work best with numerical data. However, categorical variables (e.g., "City", "Brand", "Day") are often stored as text (strings). Since ML algorithms cannot directly process text-based categories, we must convert them into a numerical formatâ€”this is where dummy variables (one-hot encoding) come in.

What is One-Hot Encoding?

->One-hot encoding transforms categorical values into binary (0 or 1) columns, creating a new column for each category.

**Why is One-Hot Encoding Needed?**

1.   Machine Learning Algorithms Require Numeric Data

     ->Many models (e.g., Linear Regression, Logistic Regression, SVM, Neural Networks) work only with numbers. One-hot encoding helps make categorical data usable.
2.   Avoids Misinterpretation of Categorical Values

    ->If categories are encoded as numbers (e.g., "Apple" = 1, "Carrot" = 2), models may assume an incorrect order or hierarchy, which is misleading.


3.   Preserves Data Information

  ->Unlike label encoding (which assigns arbitrary numbers to categories), one-hot encoding ensures that the model treats categories independently.

2.   Improves Model Performance

 ->It allows models to learn relationships more accurately by expanding categorical features into separate binary columns.













**Apply one-hot encoding to the categorical columns**

```python
categorical_columns2 = df.select_dtypes(include=['object', 'category']).columns
```

```python
# Apply one-hot encoding
df_one_hot = pd.get_dummies(df, columns=categorical_columns2, drop_first=True)  # Avoid redundancy
```

```python
# Save the transformed dataset
output_file = "sales_data_transformed.csv"
df_one_hot.to_csv(output_file, index=False)
```
